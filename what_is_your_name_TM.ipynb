# Can 모델 생성

```python
from keras.models import load_model
from PIL import Image, ImageOps
import numpy as np

# Disable scientific notation for clarity => # 소수지수표현(과학적표기법)을 일반식으로 변경
np.set_printoptions(suppress=True)
# Load the model
model = tensorflow.keras.models.load_model('/content/drive/MyDrive/해커톤_공유폴더/t.m_model/keras_model_CAN.h5')
```

```python
model.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
# 예측에 사용할 데이터(validation data)
test_datagen = ImageDataGenerator(rescale=1./255)

test_generator = test_datagen.flow_from_directory(
        PATH + 'can_mini/test',
        target_size=(224, 224),    
        # batch_size=3,
        class_mode='categorical', shuffle=False, seed=0)

# 모델 평가하기
print()
print("-- Evaluate --")
scores = model.evaluate_generator(test_generator)
print("%s: %.2f%%" %(model.metrics_names[1], scores[1]*100))

# 모델 예측하기
print()
print("-- Predict --")

output = model.predict_generator(test_generator)
np.set_printoptions(formatter={'float_kind': lambda x: "{0:0.3f}".format(x)})

print(output)

for s in range(0, len(test_generator.filenames)):
  print(test_generator.filenames[s])
```
